# **EXPERIMENT: 2**
 ## Measures of Central Tendency with Visualization
 **Aim:**

To demonstrate the calculation of basic descriptive statistics, specifically measures of central tendency (mean, median, and mode), for a
given dataset.
Objectives:
To calculate and display the mean, median, and mode of a given dataset of scores.

**Algorithm:**

Step 1:Import the statistics module

Step 2:Define the dataset

Step 3:Calculate the descriptive statistics

Step 4: Display the results

**CASE:**
Calculate the measures of central tendencies for a given scores:
12, 15, 18, 20, 22, 15, 18, 25, 30, 18

```python
# GIVEN DATA
data = [12, 15, 18, 20, 22, 15, 18, 25, 30, 18]
print("Given Data:", data)
GivenData: [12, 15, 18, 20, 22, 15, 18, 25, 30, 18]
import statistics
# Mean
mean_score = statistics.mean(data)
# Median
median_score = statistics.median(data)
# Mode
mode_score = statistics.mode(data)
# Display the results
print("Mean:", mean_score)
print("Median:", median_score)
print("Mode:", mode_score)
```

```python
# VISUALIZATION
# HISTOGRAM – Data Distribution
import matplotlib.pyplot as plt
plt.figure()
plt.hist(data, bins=6)
plt.xlabel("Data Values")
plt.ylabel("Frequency")
plt.title("Histogram of Given Data")
plt.show()
# BOXPLOT – Spread and Central Value
plt.figure()
plt.boxplot(data, vert=False)
plt.xlabel("Data Values")
plt.title("Boxplot of Given Data")
plt.show()
```

```python
# INTERPRETATION:
# - Histogram shows frequency distribution of data
# - Boxplot shows median, range, and possible outliers
# - Mean is affected by extreme values
# - Median gives the central value
# - Mode shows the most repeated observation
# ------------------------------------------------------------
# CONCLUSION:
# Statistical measures along with visualization provide
# a clear understanding of the data distribution and
# central tendency in engineering applications.
```

**Result and Discussion:** The code effectively calculates and displays the mean, median, and mode of the given dataset, providing
valuable information about the central tendencies of the values. These results can be used for basic statistical analysis and initial
interpretations of the data.

# **Application Problem 2:**
Analysing CPU Load Consistency

Problem Statement

A computer engineer is analysing the CPU load values of a server over 30 monitoring intervals in a data center. Each interval was recorded
under controlled operating conditions to measure the processor load (in percentage). The observed CPU load values are:

28.5, 30.2, 29.8, 31.1, 28.9, 30.0, 30.4, 29.9, 31.2, 30.6, 28.7, 29.5, 30.3, 29.7, 30.8, 30.1, 30.5, 30.0, 29.3, 31.0, 28.8, 30.9, 29.6, 30.2, 31.3, 30.0,
29.4, 30.7, 28.6, 30.3.

The objective is to statistically analyse the results to evaluate the average CPU load, the most frequent load value, and the middle load
value when the data is sorted, in order to assess the stability and reliability of the system.


```python
cpu_load = [ 28.5,30.2, 29.8, 31.1, 28.9, 30.0, 30.4, 29.9, 31.2, 30.6, 28.7, 29.5, 30.3, 29.7,
30.8, 30.1, 30.5, 30.0, 29.3, 31.0, 28.8, 30.9, 29.6, 30.2, 31.3, 30.0, 29.4, 30.7, 28.6, 30.3]
import statistics
# Mean
mean_load = statistics.mean(cpu_load )
# Median
median_load = statistics.median(cpu_load )
# Mode
mode_load = statistics.mode(cpu_load)
# DISPLAY RESULTS
print("CPU Load Analysis")
print("-----------------")
print("Mean CPU Load :", mean_load)
print("Median CPU Load :", median_load)
print("Mode CPU Load :", mode_load)
# VISUALIZATION
# HISTOGRAM – Data Distribution
import matplotlib.pyplot as plt
plt.hist(cpu_load, bins=8)
plt.xlabel("CPU Load (%)")
plt.ylabel("Frequency")
plt.title("Distribution of CPU Load")
plt.show()
```
