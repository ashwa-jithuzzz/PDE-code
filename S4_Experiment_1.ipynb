4. SciPy – Probability Distributions
(a) Binomial Distribution

```python
from scipy.stats import binom ### binom represents the Binomial Distribution.
##scipy.stats → statistics module
n = 10 # number of trials
p = 0.5 # probability of success
x = 4 # number of successes
prob = binom.pmf(x, n, p)
print("P(X = 4) =", prob)
```

(b) Poisson Distribution

```python
from scipy.stats import poisson
lam = 3 # mean rate
k = 2 # number of events
prob = poisson.pmf(k, lam)
print("P(X = 2) =", prob)
```

(c) Normal Distribution

```python
from scipy.stats import norm
mean = 50
std = 5
prob = norm.cdf(55, mean, std)
print("P(X ≤ 55) =", prob)
```

5. Matplotlib – Visualization

(a) Histogram of Data
A histogram is a graphical representation that shows how data values are distributed
over a range.

X-axis → Data values

Y-axis → Frequency (count of values)

In probability and statistics, histograms help us:
Understand data distribution
Identify patterns (normal, skewed, uniform)
Relate experimental data to probability distributions

bins=10

Bins divide the entire data range into equal intervals.
Meaning of bins=10: The range of data is split into 10 intervals (groups).
Each bin counts how many values fall in that interval.

Example:

If data range is 10 to 30:
Bin width = (30 - 10) / 10 = 2

```python
import matplotlib.pyplot as plt
data = [12, 15, 18, 20, 22, 25, 30]
plt.hist(data, bins=12)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Data")
plt.show()
```

(b) Plotting Binomial Distribution

```python
x = np.arange(0, 11)
y = binom.pmf(x, n=10, p=0.5)
plt.stem(x, y)##plt.stem() is used to draw a stem plot, which is mainly used to
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.title("Binomial Distribution")
plt.show()
```

(c) Plotting Normal Distribution Curve

```python
x = np.linspace(30, 70, 100)
y = norm.pdf(x, loc=50, scale=5)
##loc → mean (μ), scale → standard deviation (σ)
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.title("Normal Distribution Curve")
plt.show()
```

6. **Statsmodels**
Statistical modeling and econometrics.

```python
import statsmodels.api as sm
X = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
model.summary()
```
